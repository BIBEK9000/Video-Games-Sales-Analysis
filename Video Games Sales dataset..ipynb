```python
#question 1) from video games dataset.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
games_df = pd.read_csv(r'E:\pandas iris dataset\vgsales.csv')
games_df.head(10)
```

```python
#question 2) from video games dataset.
#Here i take "total_publishers" as a variable on that i put my dataset with specific "Publisher" column with .nunique() to get the number of unique elements series or column of a dataframe it will count that specific column and also get total number of unique that i specified atlast i print it. 
total_publishers = games_df["Publisher"].nunique()
print("Total number of unique publisher:" , total_publishers)
```

```python
#question 3) from video games dataset.
#Here i take "sports_df" as a variable and then i filter the my dataframe and take "Genre" column is equal to sports that includes only rows and then take total_sales variable on that i put sports_df variable along with "Global_sales" column with the summing values function .sum() atlast i print it.  
sports_df = games_df[games_df["Genre"] == "Sports"]
total_sales = sports_df["Global_Sales"].sum()
print("Global sales for the sports genre:" , total_sales)
```

```python
#question 4) from video games dataset.
#Here i take a variable "filtered_df" on that i'm filtering that my given dataset to column for "Year" >= 2010 & operator to combine other column "Platform" == 'PS4' atlast i print it.
filtered_df = games_df[(games_df["Year"] >= 2010) & (games_df["Platform"] == "PS4")]
print("year column platform column:" ,filtered_df["Year"] ,["Platform"])
```

```python
#question 5) from video games dataset.
#Here i take a variable as "genre_sales" on that i put my dataset that's games.df and using .groupby() function overthat i grouped "Genre" column along  with "Global_Sales" column with .mean() to find average between that atlast i print it.
genre_sales = games_df.groupby("Genre")["Global_Sales"].mean()
print(genre_sales)
```

```python
#question 6) from video games dataset.
#Here i took a variable as "publisher_sales" on that use my dataset using with .groupby() function along with "Publisher" column with "Global_Sales" column and using .sum() function to sum of both the columns afterthat i took a another variable as "top_publishers" on that i put my "publisher_sales" variable with .nlargest(5) function and put the parameter 5 on it so i can get top 5 publishers atlast then print it........ here i can't use directly .max() function coz .max() function don't have any parameter like this if i put .max() function i do some extra logic i can took avariable overthat combine both the variable == then using .head(5) function i can put the parameter . 
publisher_sales = games_df.groupby("Publisher")["Global_Sales"].sum()
top_publishers = publisher_sales.nlargest(5)
print(top_publishers)
```

```python
#question 7) from video games dataset.
#Here i do first all the selected columns summing and store the result in a new column called "Total_Sales" and then we group the dataframe by "Platform" and calculate the mean of "Total_Sales" for each platform using .groupby() function and finally sort the average_sales_by_platform using .sortvalues() fucntion along with parameter ascending=False that represnts in descending order and then print it.
games_df["Total_Sales"] = games_df["NA_Sales"] + games_df["EU_Sales"] + games_df["JP_Sales"] + games_df["Other_Sales"]
avg_sales_by_platform = games_df.groupby("Platform")["Total_Sales"].mean()
avg_sales_sort_platform = avg_sales_by_platform.sort_values(ascending=False)
print(avg_sales_sort_platform)
```

```python
#question 8) from video games dataset.
#here i do first two selected columns and summarise of two columns on that  stored a result in a new column "Total_Sales_globally" and then using .groupby() function of my dataframe groupby "Year" "total_sales_globally" along with .sum() function and then used .idxmax() function to get the highest values of the year and then filtering the "Year" column == on that stored variable "year_with_highest_sales" afterthat we calculate top_3_games with .nlargest() fucntion along with parameter (3 , "Total_Sales_globally") and specific column i already did atlast print it "year" and "Total_Sales_globally" column .  
games_df["Total_Sales_globally"] = games_df["Total_Sales"] + games_df["Global_Sales"]
total_global_sales_year = games_df.groupby("Year")["Total_Sales_globally"].sum()
year_with_highest_sales = total_global_sales_year.idxmax()
games_highest_sales_year = games_df[games_df["Year"] == year_with_highest_sales]
top_3_games = games_highest_sales_year.nlargest(3 , "Total_Sales_globally")
print(top_3_games[["Year" , "Total_Sales_globally"]])
```

```python
#question 9) from video games dataset.
#Here i take a variable as a "sorted_df" in my existing dataframe i use .sort_value() fucntion over existing column "Global_Sales" in a descending order on that variable i already took "sorted_df" i use .reset_index() function because we need to reset index of the sorted dataframe and set oarameter 'drop=True' removes orginal index column and replace it with anew sequential starting from 0 this ensure correct ranking games based on the "Global_Sales" without resetting the index values would still correspond to the orginal dataframe then we have to problem with findidng sorted values and also use parameter "inplace=True" so we can modify the dataframe on my existing dataframe if "inplace=False" then will show a completely a new dataframe afterthat i put a new column"Sales_Rank" with the "sorted_df" variable and use .index() +1 to incrament by one by one order afterthat i took a new variable "games_df" using pandas as pd and merge with my existing dataframe along with new dataframe where "Name"  column is the key and left join used to merge "Sales_Rank" and also use the parameter 'on=Name' merging operation should be based on the "Name" column and 'how=Left' the merging opeartions should be left join in a left join it will merge from left dataframe "games_df" and merege  from right dataframe "sorted_df" if a row in left and right dataframe does not have a match then it will give a MISSING VALUES in the merged datafame atlast i print it. 
sorted_df = games_df.sort_values("Global_Sales" , ascending=False)
sorted_df.reset_index(drop=True , inplace=True)
sorted_df["Sales_Rank"] =sorted_df.index+1
df = pd.merge(games_df, sorted_df[["Name" ,"Sales_Rank"]], on="Name" , how="left")
print(df[["Sales_Rank"]])
```

```python
#question number 10) from video games dataset.
#Here i took a variable called 'region_sales' on that i grouped in my existing dataframe and groupedby "Genre" columns with other columns then .sum() function to get the sum of all functions and then took a variable called result_df on that a new dataframe i create overthat i put a variable "region_sales" overthat atlast i print it. 
region_sales = games_df.groupby("Genre")["NA_Sales" , "EU_Sales" , "JP_Sales" ,"Other_Sales"].sum()
result_df = pd.DataFrame(region_sales)
print(result_df)
```

```python
#question number 11) from video games dataset.
#Here i took a variable named as "platform_counts" on that in my existing dataframe i took specific column "Platform" with .value_counts() method function to get the unique values in the series and then use seaborn function that i want a barplot and the parameter x=.index function to indexing from x and y=.values function to get values from y and labels and title with the plot and also use xticks() function to rotation from x degree atlast plt.show() function to visualize.  
platform_counts = games_df["Platform"].value_counts()
sns.barplot(x=platform_counts.index , y=platform_counts.values)
plt.xlabel("Platform")
plt.ylabel("Number of games released")
plt.title("Number of games released for each platform")
plt.xticks(rotation=90)
plt.show()
```

```python
#question number 13) from video games dataset.
#Here i use seaborn libaries to boxplot on that "Genre" column on x-axis and "EU_Sales" on y-axis we also pass the dataframe "games_df" parameter data to specify the data source and labeling in x and y axis and title and use xticks on x-axis and roration to avoid overlapping atlast plt.show() to visualize.  
sns.boxplot(x="Genre" , y="EU_Sales" , data=games_df)
plt.xlabel("Genre")
plt.ylabel("Eu Sales")
plt.title("Eu Sales Distribution by Genre")
plt.xticks(rotation=90)
plt.show()
```

```python
#question number 14) from video games dataset.
#Here i first doing that we are converting the "Year" column of my dataframe to datetime data type using pd.to_datetime() function is used to convert the values in the "year" column to a datetime object which allows us to perform time-based operations on th data that pd.to_datetime(games_df["Year"] , format=%y) this function converts the values in the "year" column to datetime format specified '%Y' the format specifies that the values in the "Year" column are reoresented as fouur-digit year (2010,2011) afterthat i used dt.year attribute to extracts the year from datetime values only "Global_Sales" part selects only the "Global_Sales" column grouped data overthat .mean() is used for average and also .reset_index() function is used converting the grouped year that previously set as the index back to a regular column after excuting this line Average_Sales  dataframe will contain two columns "Year" "Global_Sales" along with "Global_Sales" column represnting the average global sales for games released in each year afterthat using seaborn and matplotlib i did style ,lineplot(i took x=year , y=average sales ,data="average_sales" variable ,labels ,title and last show  to visualize my line plot.                  
games_df["Year"] = pd.to_datetime(games_df["Year"] , format="%Y")
average_sales = games_df.groupby(games_df["Year"].dt.year)["Global_Sales"].mean().reset_index()
sns.set_style("darkgrid")
plt.figure(figsize=(10,6))
sns.lineplot(x="Year" , y="Global_Sales"  , data=average_sales)
plt.xlabel("Year")
plt.ylabel("average Global SaLES")
plt.title("average global sales of the year")
plt.show()
```

```python
#question number 15 from video games dataset.
#Here i took a variable sales_columns on that i put all specific sales columns in my dataframe and then i use for loop on that i plot a figure size 10inches is width and 8inches is length afterthat i use seaborn to histogram plot so i use histplot() function where parameter kde=True means kernel density estimation curve to the histogram for smooth representation of the distribution and also use color parameter afterthat i give plt.title() function this line sets the title specifying sales column and usinf f-string to allow with the values or columnsname and then label to x-axis of the plot to the current sales of the column name and y-axis labels and show to visualize  
sales_columns = ["NA_Sales" , "EU_Sales" , "JP_Sales" , "Other_Sales" , "Global_Sales"]
for column in sales_columns:
    plt.figure(figsize=(10,8))
    sns.histplot(games_df[column] , kde=True , color="skyblue")
    plt.title(f"distribution of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
```
